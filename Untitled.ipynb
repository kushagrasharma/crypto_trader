```python
data.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Mid': 'mid', 'Bid': 'bid', 'Ask': 'ask', 'Last': 'close', 'Volume': 'volume'}, inplace=True)
data
```

```python
def init_state(data):
#     high = data["high"].values
#     low = data["low"].values
#    mid = data["Weighted_Price"].values
#     last = data["Last"].values
#     bid = data["Bid"].values
#     ask = data["Ask"].values
#     volume = data["Volume"].values
    
 #   priceDiffs = np.diff(mid)
 #   priceDiffs = np.insert(priceDiffs, 0, 0)
    
    # William's %R
    willr = WILLR(data, timeperiod=14)
    # Double Exponential Moving Average
    dema = DEMA(data, timeperiod=30)
    # Chaikin A/D Oscillator
    chaikinOscillator = ADOSC(data, fastperiod=3, slowperiod=10)
    # Chaikin A/D Line
    chaikinLine = AD(data)
    # Normalized Average True Range
    trueRange = NATR(data, timeperiod=14)
    # Linear Regression Slope
    linearRegSlope = LINEARREG_SLOPE(data, timeperiod=14)
    # Standard Deviation
    std = STDDEV(data, timeperiod=5, nbdev=1)
    # Time Series Forecast
    tsf = TSF(data, timeperiod=14)
    
    xdata = np.column_stack((willr, dema, chaikinOscillator, chaikinLine, trueRange, linearRegSlope, std, tsf))
    return xdata
```

```python
states = init_state(df)
```
